```python
# problem 120
# very slow
# better math needed https://projecteuler.net/thread=120

def reminder120(start, end):
    sumr = 0
    maxes = {}
    for a in xrange(start, end+1):
        maxr = 0
        maxrn = 0
        # remainder cycles are at most 2a long
        for n in xrange(1, 2*a+1):
            r = ( pow((a-1), n) + pow((a+1), n) ) % a**2 
            if r > maxr: 
                maxr = r
                maxrn = n
        sumr += maxr
        maxes[a] = (maxrn, maxr)
    return sumr, maxes
```
